# Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/briankimanzi/Downloads/LOTR.csv')
df1
```

```python
df2 = pd.read_csv('/Users/briankimanzi/Downloads/LOTR 2.csv')
df2
```

```python
df1.merge(df2, how= 'inner' , on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how= 'outer')
```

```python
df1.merge(df2, how= 'left')
```

```python
df1.merge(df2, how= 'right')
```

```python
df1.merge(df2, how= 'cross') # takes values from right data frame and compares the value to the left data frame 

```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
# Concatenate
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```

# Pandas Data cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/briankimanzi/Downloads/Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# df = df.drop(columns = 'Not_Useful_Column') # drops the un needed column
df
```

```python
# df['Last_Name'].str.lstrip("...")
# df['Last_Name'] = df['Last_Name'].str.lstrip("/")
# df['Last_Name'] = df['Last_Name'].str.lstrip("...")
# df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df['Last_Name'] = df['Last_Name'].str.lstrip("123._/")

# df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex= True)
# df['Phone_Number']= df['Phone_Number'].apply(lambda x : str(x))

# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '', regex = True)
# df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '', regex = True)
# df
df
```

```python
df[['street Address', "state", "zip_code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y', regex = True)

df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N', regex = True)

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y', regex = True)
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'Y', regex = True)
df
```

```python
# df.replace('N/a', '', regex = True)
df = df.fillna('')
# df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

# df  = df.dropna(subset="Phone_Number", inplace= True) # Another way to drop null values
# df
```
